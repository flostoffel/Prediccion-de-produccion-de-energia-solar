```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sunlab-faro-meteo-2017.csv", sep=";")
df =df.sort_values("Datetime").reset_index(drop=True)
l_meteo=list(df["Datetime"])
l4=[]
for i in l_meteo:
    dates=i[:19]
    dates_time =datetime.strptime(dates, "%Y-%m-%dT%H:%M:%S")
    l4.append(dates_time)
l5 =pd.Series(l4)
df["Datetime"]=l5
```

```python
df1 = pd.read_csv("sunlab-faro-pv-2017.csv", sep=";")
df1 =df1.sort_values("Datetime").reset_index(drop=True)
l_meteo=list(df1["Datetime"])
l4=[]
for i in l_meteo:
    dates=i[:19]
    dates_time =datetime.strptime(dates, "%Y-%m-%dT%H:%M:%S")
    l4.append(dates_time)
l5 =pd.Series(l4)
df1["Datetime"]=l5
```

```python
df=df.merge(df1, on ="Datetime")
df.head()
```
